```python
#Inheritance
#New class inherits another
#A child class from a Parent class
#Takes attrs from Parent class

class Car:
    def __init__(self, manufacturer, model, year):
        self.manufacturer = manufacturer
        self.model = model
        self.year = year
        self.odometer = 0
    
    def description(self):
        name = f"{self.manufacturer} {self.model} {self.year}"
        print(name)
        
    def odometer_read(self):
        print(f"This car has recorred {self.odometer} miles")
    
    def fillTank(self):
        print('The tank is filling')
 
 #Child Class
#The Subclass takes the Car class subclass(ParentClass)
#Electric Car is a type of Car that can take its attrs but has special characteristics
class ElectricCar(Car):
    def __init__(self, manufacturer, model, year):
        super().__init__(manufacturer, model, year) #super() allows to call a method from the parent class superclass - subclass
        self.battery = 21 #adding new characteristics to it
    
    def batteryPercentage(self):
        print(f"Car has {self.battery}% of battery")
    
    def fillTank(self): #Overriding the method
        print("This car doesn't use a tank")

eCar = ElectricCar('Tesla', 'A20', 2020)
eCar.description()
eCar.batteryPercentage()

#Override methods from the parent class
eCar.fillTank() #We can override methods for the car, in the child class
```

```python
#Ejs
class Restaurante:
    def __init__(self, nombre, cocina):
        self.nombre = nombre
        self.cocina = cocina
        self.number_served = 0
    def descripcion(self):
        print(f"El restaurante '{self.nombre.title()}', sirve {self.cocina}")
    def abierto(self):
        print(f"{self.nombre}, está abierto\n")
    def numberServed(self):
        print(f"El resturante {self.nombre} ha atendido a {self.number_served} personas")
    def set_numberServed(self, newValue):
        self.number_served = newValue
    def increment_numberServed(self, Incrementvalue):
        self.number_served += Incrementvalue
        
class PuestodeHelados(Restaurante):
    def __init__(self, nombre, cocina="Helado"):
        super().__init__(nombre, cocina)
        self.sabores = ["Chocolate", "Fresa"]
        
    def descripcion(self):
        print(f"El restaurante {self.nombre} sirve buenos helados")
        
    def listaHelados(self):
        print("\nTiene helados disponibles:\n")
        for sabor in self.sabores:
              print(f"* {sabor}")
        
mirestaurante = PuestodeHelados("Barbaros")
mirestaurante.descripcion()
mirestaurante.listaHelados()

mirestaurante = Restaurante("Barbaros", "Comida Italiana")
#mirestaurante.descripcion()
print(f"\nEl restaurante {mirestaurante.nombre}, también sirve buena {mirestaurante.cocina}")
```

```python
class Users:
    def __init__(self, firstName, lastName, age):
        self.firstName = firstName
        self.lastName = lastName
        self.age = age
        self.login_attepmts = 0
    def description(self):
        print(f"\nFirstname: {self.firstName}\nLastname: {self.lastName}\nAge: {self.age}")
    def greet(self):
        print(f"\nHi {self.firstName} {self.lastName}!!")
    def increment_login_attempts(self):
        self.login_attepmts += 1
        print(f"Login attempts: {self.login_attepmts}")
    def reset_login_attempts(self):
        self.login_attepmts = 0
        print(f"\nLogin attempts: {self.login_attepmts}")
        

class Privileges:
    def __init__(self):
        self.privileges = ["Can delete post", "Can add post", "Can ban user"]
    
    def showPrivileges(self):
        print("Your privileges are:\n")
        for privilege in self.privileges:
            print(f"* {privilege}") 

class Admin(Users):
    def __init__(self, firstName, lastName, age):
        super().__init__(firstName, lastName, age)
        self.privileges = Privileges()

            
admin1 = Admin('Johan', 'Posada', 16)
admin1.greet()
admin1.privileges.showPrivileges()

admin2 = Admin("Elliot", "Alderson", 25)
admin2.greet()
admin2.privileges.privileges = ["Can do anything"] #Modificando el attr
admin2.privileges.showPrivileges()
```
