```python
#Lectura de Archivos
fileManager = open('mbox-short.txt')
count = 0

#open no puede leer el archivo por la poca memoria, gigabytes
#podemos saber la cantidad de líneas del archivo con un bucle for, este cuenta el salto de línea
#se imprimen los datos linea a linea, cada línea que es leída se descarta, sin quedarse corto de memoria  
for line in fileManager:
    count = count + 1
    print('La cantidad de líneas es: ', count)
```

```python
#Búsqueda a través de un archivo
#ignorar líneas, procesar las que nos interesan, con métodos de cadenas
#startswith, para elegir las cadenas con el prefijo

fileManager = open('mbox-short.txt')
count = 0
for line in fileManager:  #print(dir(line))
    if line.startswith('From:'):
        print(line)  
```

```python
#La cadena almacenada en line, incluye saltos de línea, estos se pueden eliminar con rstrip

fileManager = open('mbox-short.txt')
for line in fileManager:
    line = line.rstrip()
    if line.startswith('From:'):
        print(line)
```

```python
fileManager = open('mbox-short.txt')
for line in fileManager:
    line = line.rstrip()
#ignorar linea que no nos interesa, continue la salta
    if not line.startswith('From:'):
        continue 
#la línea que si nos interesa    
    print(line)
```

```python
#encontrar lineas donde el string(cadena) aparece
fileUser = input('Inserta el nombre del archivo: ')
fileManager = open(fileUser)
for line in fileManager:
    line = line.rstrip()
#Si la cadena no fue encontrada, es decir nos devuelve un "-1", lo salta con continue
    if line.find('@uct.ac.za') == -1:
        continue
    print(line)
```

```python
#Escritura de archivos

file = open('romeo.txt', 'w')
print(file)

#Si el archivo existía, abrirlo en modo escritura hara que todo se borre, si no existía, uno nuevo se creará
#usamos "var.write()" para escribir en el archivo, este nos devulve la longitud de los caracteres escritos
_1line = 'Here I am, \n'
_2line = 'Oh I didn''t saw you'
print(file.write(_1line))
print(file.write(_2line))
#cerramos el archivo
file.close()
```
