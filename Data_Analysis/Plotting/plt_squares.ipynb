```python
import matplotlib.pyplot as plt

#Data that we'll plot
squares = [1, 4, 9, 16, 25]

#subplots() function can generate one or more plots in the same figure
#fig: represents the entire figure or collection of plots that are generated
#ax : represents a single plot in the figure, used it the most of the time

fig, ax = plt.subplots()
ax.tick_params(axis='x', colors='red')    #setting up X-axis tick color to red
ax.tick_params(axis='y', colors='red')

#If we give to the plot() a sequence of numbers, it assumes the 1st point corresponds to 0 in y axis
ax.plot(squares)

plt.show()
```

```python
#Changing the Label Type and Line Thickness

import matplotlib.pyplot as plt

#Data that we'll plot
input_values = [1, 2, 3, 4, 5]
squares = [1, 4, 9, 16, 25]

fig, ax = plt.subplots()

#Setting the x values and the corresponding y values (correction).
ax.plot(input_values, squares, linewidth=3)

#Setting the chart tittle and the label axes
ax.set_title("Números al Cuadrado,", fontsize=24)

#Title for each of the axes
ax.set_xlabel("Valor" , fontsize=14)
ax.set_ylabel("Cuadrado del valor", fontsize=14)
ax.set_yticks(range(0, 25, 2))

#tick_params styles the tick marks, we indicate the axes to aply it, and tick mark label size
ax.tick_params(axis="both", labelsize=14)

plt.show()
```
