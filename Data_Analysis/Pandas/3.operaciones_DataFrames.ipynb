````python
data = np.random.randint(21, size=10)

df = pd.DataFrame(data)
df
````

### Añadiendo una nueva columna

````python
df['new_col'] = 21
df['profesión'] = 'programador'
df
````

````python
df['x * e'] = [(x * np.e)for x in range(10)]
df
````

````python
df['(x * e) * 2'] = df['x * e'] * 2
df
````

### Cambiando el nombre de columnas

````python
df.columns = ['n', 'c', 'profesion', 'x * e', df.columns[4]]
df
````

````python
# Usando el metodo rename()
df.rename(columns= {'c' : 'CONSTANT'}, inplace=True)
df
````

````python
df.rename(columns= {'n' : 'number', 'x * e' : 'i * e', df.columns[4] : f"({df.columns[3]}) * 2"}, inplace=True)
df
````

### Modifcando una columna

````python
df['CONSTANT'] = df['CONSTANT'] / 3
df
````

### Eliminando una columna o fila

````python
# Usando del
del df['(x * e) * 2']
df
````

### Descartando filas y columnas

* El parámetro 

```python  
  inplace=True indica que el descarte (drop) va a ser permanente

````python
# Usando drop()   drop = descartar
#Borrando filas (axis=0) específicas


#En este caso no se usa el parámetro inplace=True 
clear = df.drop([0 ,1 , 3], axis=0) 
clear
````

````python
#Aquí se muestra que el descarte no fue permanente
df
````

````python
#Usando (inplace=True) para hacer los cambios permanentes

df.drop(['number'], axis=1, inplace=True)
df.head()
````
