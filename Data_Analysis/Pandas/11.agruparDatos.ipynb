````python
import pandas as pd
import numpy as np
````

````python
df = pd.read_csv("data_celular.csv",
                index_col=0, 
                names=['index', 'Fechas', 'Duración', 'Item', 'Mes',  'Red', 'Tipo_red'], 
                parse_dates=['Fechas'],
                header=0)

#header=0 establece la primera fila como cabecera
#names: renombra las columnas
#La duración se mide en segundos
#Fechas hace referencia a cuando se creó la entrada en el archivo de datos
df.head()
````

### Aplicando operaciones al df para obtener info. básica

````python
qFilas = df['Item'].count()
print(f"El df tiene {qFilas} Filas")
````

````python
i = df['Duración'][df['Item'] == 'call']
i.head()
````

````python
segundosTotal = (df['Duración'][df['Item'] == 'call']).sum() #Sumar la duracion donde item == 'call'
print(f"El tiempo total para las llamadas fue de: {segundosTotal} segundos")
````

````python
#Calculando la cantidad de redes contactadas de 2014/11(mes inicial) a 2015/03(mes final)

cantidadRedes = df['Red'].nunique() #.nunique() cuenta la cantidad de datos unicos (un dato repetido como [5, 5, 5] lo cuenta como unico)
print(f"La cantidad de redes contactadas de (2014-11 a 2015-03) fue de: {cantidadRedes}")
````

### ```.groupby()```

* Permite agrupar datos para obtener info. segregada

````python
grouping = df.groupby('Red')
grouping.groups.keys()
#grouping.groups.values()
````

````python
grouping.first()
````

````python
df.groupby('Mes').sum()
````

````python
print("Cantidad de llamadas por mes segregadas en llamadas, datos y sms")
df.groupby(['Mes', 'Item'])["Duración"].count() #Agrupando por Mes e Item y de los datos agrupados hallando la cantidad de llamadas por mes de call, data, sms
````

````python
df.groupby('Mes').first()
````

````python
df[df['Item'] == 'call'].groupby('Red').first()
````

````python
print('La cantidad de llamadas realizadas a cada una de las operadores es:')
df[df['Item'] == 'call'].groupby('Red')['Duración'].sum()
````

````python
print('Cuantas veces son contactadas cada operadora por mes para llamadas, datos y sms')
cantidad = df.groupby(['Mes', 'Tipo_red'])['Duración'].count()
cantidad = df.groupby(['Mes', 'Tipo_red'])['Fechas'].count() #Al contar cuanquier columna, el resultado va a ser el mismo
cantidad
````
