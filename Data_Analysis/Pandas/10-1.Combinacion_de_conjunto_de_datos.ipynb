````python
#Mirando en que filas de la columna NAME está la estación de MASSENA
precipitaciones['NAME'] == "MASSENA INTERNATIONAL AIRPORT, NY US"
````

````python
#Filtrando en donde MASSENA aparece (columnas del df en donde la condicion se cumple)
precipitaciones[precipitaciones['NAME'] == "MASSENA INTERNATIONAL AIRPORT, NY US"]
````

````python
precip_massena = precipitaciones[precipitaciones['NAME'] == "MASSENA INTERNATIONAL AIRPORT, NY US"]
precip_massena.shape
````

### Usando ```.merge()``` para hacer *joins*  entre DataFrames

* por defecto la función hace un *Inner Join* (Intersección)

````python
massena_inner_merge = pd.merge(precip_massena, temperaturas)
massena_inner_merge
````

````python
massena_outer_merge = pd.merge(precip_massena, temperaturas, how="inner", on=["STATION", "DATE"])
massena_outer_merge.columns
````
